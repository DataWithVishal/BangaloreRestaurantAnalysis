### Import the Dependencies


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the dataset in zomato_df 

```python
zomato_df = pd.read_csv('zomato.csv')
zomato_df.head()
```

```python
zomato_df.info()
```

```python
zomato_df.shape
```

```python
zomato_df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'], axis = 1,inplace=True)
zomato_df.head()
```

```python
zomato_df.info()
```

```python
zomato_df.isnull().sum()
```

```python
zomato_df.drop_duplicates(inplace=True)
zomato_df.info()
```

```python
zomato_df['listed_in(type)'].value_counts()
```

```python
zomato_df['listed_in(city)'].value_counts()
```

```python
# clean rate column
def cleanrate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
zomato_df['rate'] = zomato_df['rate'].apply(cleanrate)
zomato_df.head()
```

```python
# Handle the null Values
zomato_df['rate'] = zomato_df['rate'].fillna(zomato_df['rate'].mean())
zomato_df.dropna(inplace=True)
zomato_df.head()
```

```python
 zomato_df.info()
```

```python
# Rename Columns
zomato_df.rename(columns={'approx_cost(for two people)':'Cost4two','listed_in(type)':'Type','listed_in(city)':'City'},inplace=True)
zomato_df.info()
```

```python
def intconvert(value):
    value = str(value).replace(',','')
    return float(value)
zomato_df['Cost4two'] = zomato_df['Cost4two'].apply(intconvert)
zomato_df.head()
```

```python
locations = zomato_df['location'].value_counts()
locationlessthan100 = locations[locations<100]
def handle_location(value):
    if (value in locationlessthan100):
        return 'others'
    else:
        return value
zomato_df['location'] = zomato_df['location'].apply(handle_location)
zomato_df['location'].value_counts()
```

```python
cuisines = zomato_df['cuisines'].value_counts()
cuisineslessthan50 = cuisines[cuisines<50]
def handle_cuisines(value):
    if (value in cuisineslessthan50):
        return 'others'
    else:
        return value
zomato_df['cuisines'] = zomato_df['cuisines'].apply(handle_cuisines)
zomato_df['cuisines'].value_counts()
```

```python
rest_type = zomato_df['rest_type'].value_counts()
rest_typelessthan100 = rest_type[rest_type<100]
def handle_rest_type(value):
    if (value in rest_typelessthan100):
        return 'others'
    else:
        return value
zomato_df['rest_type'] = zomato_df['rest_type'].apply(handle_rest_type)
zomato_df['rest_type'].value_counts()
```

## Visualization

```python
plt.figure(figsize = (16,10))
ax = sns.countplot(data=zomato_df,x='location')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14,8))
sns.boxplot(x='Type',y='rate',data = zomato_df)
plt.show()
```

```python
plt.figure(figsize=(14,8))
sns.countplot(x='book_table',data = zomato_df)
```

```python
plt.figure(figsize=(14,8))
sns.countplot(x='online_order',data = zomato_df)
```

```python
# Visualizing Online Order Data Location wise
sns.countplot(x=zomato_df.groupby('location')['online_order'].value_counts())
```

```python
order_counts = zomato_df.groupby('location')['online_order'].value_counts().unstack(fill_value=0)

# Plotting
order_counts.plot(kind='bar', stacked=False,figsize=(20,8))
plt.title('Location-wise Online Orders')
plt.xlabel('Location')
plt.ylabel('Number of Orders')
plt.legend(title='Online Order')
plt.show()
```

```python
# Average rating acording to Type of Restaurant
avg_type_rating = zomato_df.groupby('Type')['rate'].agg(np.mean)

# Plotting
avg_type_rating.plot(kind='bar', stacked=False,figsize=(20,8))
plt.title('Restaurant Type-wise Avg rating')
plt.xlabel('Type of Restaurant')
plt.ylabel('Average Rating')
plt.show()
```
